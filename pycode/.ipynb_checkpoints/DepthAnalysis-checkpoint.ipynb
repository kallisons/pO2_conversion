```python
data = xarray.open_dataset('data/TN278/os75nb.nc')
data
```

```python
# Extract amplitude and change the depth_cell to meaningful units
amp = data['amp'].T
amp.coords['depth_cell'] = data.depth.max('time')
amp = amp.rename({'depth_cell': 'depth'})
amp
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
xarray.plot.pcolormesh(amp, ax=ax)
ax.invert_yaxis();
```

```python
# resample in 10-minute intervals
amp_10min = amp.resample(freq='10T', dim='time')

fig, ax = plt.subplots(figsize=(15, 5))
xarray.plot.pcolormesh(amp_10min, ax=ax)
ax.invert_yaxis();
```

```python
amp.shape, amp_10min.shape
```

```python
# xarray does not support muliple groupbys,
# so we'll convert to a series for that
amp_df = amp_10min.to_dataframe()


# Compute time of day on a reference day
date_index = amp_df.reset_index('depth').index
timeofday = (pd.Series(date_index, index=amp_df.index)
             - pd.Series(date_index.normalize(), index=amp_df.index))
amp_df['timeofday'] = pd.to_datetime('2012-01-01') + timeofday

mean_amp = amp_df.reset_index().groupby(['depth', 'timeofday'])['amp'].mean()
mean_amp = xarray.DataArray(mean_amp.unstack())
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
xarray.plot.pcolormesh(mean_amp, ax=ax)
ax.invert_yaxis();
```

## Individual time curves

```python
from matplotlib.collections import LineCollection

def plot_time_curves(amp, ax=None, cb=True):
    if ax is None:
        fig, ax = plt.subplots()
        
    depth = amp.columns
    time = amp.index
    values = amp.values
    segments = np.array(list(np.broadcast(depth, values))).reshape(-1, len(depth), 2)
    line_segments = LineCollection(segments, alpha=0.3, cmap='viridis')
    line_segments.set_array(60 * time.hour + time.minute)

    ax.add_collection(line_segments)
    if cb:
        cb = ax.figure.colorbar(line_segments, ax=ax)
        formatter = cb.ax.yaxis.get_major_formatter()
        formatter.seq = list(map(lambda x: "{0:02}:{1:02}".format(int(x) // 60,
                                                                  int(x) % 60),
                                 formatter.seq))

    ax.axis('tight');
    ax.set(xlabel='depth', ylabel='amplitude')
    cb.set_label('time of day')


amp = data.amp.to_series().unstack()
fig, ax = plt.subplots(figsize=(12, 5))
plot_time_curves(amp['2012-04-10':'2012-04-15'], ax=ax)
```

## Modeling

```python
from scipy.optimize import minimize

def cauchy(x, gamma=1, mu=0):
    return 1. / (1 + ((x - mu) / gamma) ** 2)


def gaussian(x, sigma=1, mu=0):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2)


def model_amp(theta, depth):
    BG, Ac, gamma, Ag, mu, sigma = theta
    return BG + Ac * cauchy(depth, gamma) + Ag * gaussian(depth, sigma, mu)


def model_error(theta, depth, amp):
    return np.sum((amp - model_amp(theta, depth)) ** 2)


def param_guess(depth, amp):
    BG = amp[depth > 45].mean()
    Ac = amp[depth < 5].mean() - BG
    gamma = 6
    Ag = 40
    mu = 10
    sigma = 5
    return (BG, Ac, gamma, Ag, mu, sigma)
```

```python
amp = data.amp.to_series().unstack()

depth = amp.columns.values
vals = amp.values[2800]

result = minimize(model_error, param_guess(depth, vals), args=(depth, vals), method='BFGS')

plt.plot(depth, vals)
plt.plot(depth, model_amp(result.x, depth));
```

### Model all time steps

```python
amp = data.amp.to_series().unstack('depth_cell')

print('Hang tight... this will take a while...')
results = [minimize(model_error, param_guess(depth, amp.values[i]),
                    args=(depth, amp.values[i]), method='BFGS')
           for i in range(amp.shape[0])]
```

### Model curves

```python
model_vals = np.array([model_amp(r.x, depth) for r in results])
model_amplitude = xarray.DataArray(model_vals.T, coords=data.coords)
```

```python
amp = data.amp.to_series().unstack('depth_cell')
mamp = model_amplitude.to_series().unstack('depth_cell')

fig, ax = plt.subplots(2, figsize=(12, 10))
plot_time_curves(amp['2012-04-10':'2012-04-15'], ax=ax[0])
plot_time_curves(mamp['2012-04-10':'2012-04-15'], ax=ax[1])

ax[0].set_title('data')
ax[1].set_title('model');
```

### Model Heatmap

```python
fig, ax = plt.subplots(3, figsize=(15, 15), sharex=True, sharey=True)

xarray.plot.pcolormesh(data.amp.T, ax=ax[0])
ax[0].set_title('data')

xarray.plot.pcolormesh(model_amplitude, ax=ax[1])
ax[1].set_title('model')

xarray.plot.pcolormesh(data.amp.T - model_amplitude, ax=ax[2])
ax[2].set_title('model residuals')

ax[2].invert_yaxis();
```

```python
model_params = pd.DataFrame([r.x for r in results],
                            index=amp.index,
                            columns=['BG', 'Ac', 'gamma', 'Ag', 'mu', 'sigma'])
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
model_params.mu.plot(ax=ax, color='black', style=',')
ax.set_ylim(60, 0)
ax.set_title('Depth of the Gaussian blob');
```
